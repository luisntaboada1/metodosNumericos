***METODO DE FALSA POSICION***

(regula falsi)

```python
import numpy as np

def f(x):
    return np.exp(x) - 4 + x

rangoI = (1, 2) # Rango en x
detencion = 10**-3
xFijo = 2 # Coordenada en x fija; el punto seria (x1, f(x1)) y asi
xVar = -2
i = 0 # Contador
puntos = list()

# Punto Fijo. Este se decide previamente
pFijo = np.array([xFijo, f(xFijo)])

while True:
    pVar = np.array([xVar, f(xVar)]) # Generar el extremo de esta iteracion
    puntos.append(pVar) # Agregarlo al historial de extremos
    i = i + 1
    if abs(pFijo[0] - pVar[0]) < detencion or i > 5: # Verificar el criterio de detencion 
        break
    else: 
        xVar = pFijo[0] - ((pFijo[0] - pVar[0])/ (pFijo[1]-pVar[1]))*pFijo[1] # Siguiente iteracion

for punto in puntos:
    print(punto)
```

**GRAFICA**

```python
import matplotlib.pyplot as plt

xs = np.linspace(-2, 2, 1000)
y = (np.exp(xs) - 4 + xs)

fig, ax = plt.subplots()
grafx = []
grafy = []

# Puntos
ax.plot(xs, y)
for p in puntos:
    grafx = [pFijo[0], p[0]]
    grafy = [pFijo[1], p[1]]
    ax.plot(grafx, grafy)


# center axes at (0,0)
ax.grid(True, which="both", linestyle="--", alpha=0.5)
ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_title('f(x)')
```
