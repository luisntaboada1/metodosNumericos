***ALGORITMO DE PUNTO FIJO***

El metodo iterativo de punto fijo es un algoritmo numerico usado mas comunmente para encontrar raices de funciones de forma indirecta.
A grandes rasgos, consiste en encontrar un valor en el cual una funcion ***g(x)*** converge, cuando esta es evaluada de forma iterativa con los valores resultantes de iteraciones anteriorres, casi de forma recursiva. 

g(x0) = x1 => g(x1) = x2 => g(x2) = x3 => â€¦ => g(xn-1) = n

Esta funcion g(x) se deriva de reescribir una funcion f(x) a una expresion con la forma g(x) = x.
Como f(x) se reescribio a g(x) = x, entonces al regresar a la funcion original, los puntos que se encuentren cumplicar con f(x) = 0 (EXPLICAR DE FORMA MAS CLARA ESTA PARTE)

Por eso, lo que realmente se encutra con el algoritmo de punto fijo son los puntos en los que EL VALOR DE UNA FUNCION ES IGUAL AL DE SU ENTRADA (no modifica la entrada). Esos son los puntos fijos de una funcion

**CONDICIONES**

1. Reescribir la funcion f(x) a la forma g(x) = x
2. Que g(x) sea continua en el intervalo [a, b]
3. Que la imagen de g(x) tambien este en [a,b]
4. |g'(x)| < 1 para todos los x en [a,b] 
   - Cuando g' esta entre (0, 1), la convergencia al punto fijo es escalonada
   - Cuando g' esta entre (-1, 0) la convergencia al punto fijo es espiralada
   - Si no se cumple esta condicion, la iteracion no converge a ningun punto en especifico.


**ALGORITMO**

{DIAGRAMA DEL ALGORITMO}

Una ves que se tenga un g(x) que cumpla con los requisitos de convergencia, se aplica el algoritmo: 
1. Escoger un punto x0 inicial arbitrariamente
2. Comprobar algun criterio de detencion
3. Evaluar g(x0) = x1
4. Repetir este proceso hasta que se cumpla el criterio de detencion seleccionado (Regresar al paso 2 hasta que se cumpla)

```python
import numpy as np

# Funcion matematica
def f(x):
    return np.log(4 - x) # np.log() le aplica esa funcion a todos los elementos que se le pasen y regresa la misma cantidad de elementos

# Algortimo de punto fijo;
'''
El algoritmo regresa un diccionario con la forma {x0 : g(x0)}, osea (el valor que se evalua, el resultado de la evaluacion)
'''
def puntoFijo(func, x0, e, maxIter):
    i = 0
    iteraciones = {}
    while i < maxIter:
        i = i + 1
        x1 = func(x0)
        iteraciones[x0] = x1
        x0 = x1
    return iteraciones
```
