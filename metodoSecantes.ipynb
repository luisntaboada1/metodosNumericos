***METODO DE LAS SECANTES***

**CODIGO**

```python
import numpy as np

def f(x):
    return np.exp(x) - 4 + x


def secantes(f, x0, x1, stop, maxiter):
    """
    f : funcion matematica
    x0, x1: puntos iniciales en x
    stop, maxiter : criterios de detencion
    """
    secantes = list()
    
    i = 0
    while True:
        i = i+1 # Contar un nuevo ciclo
        p0 = (x0, f(x0)) # f en x0
        p1 = (x1, f(x1)) # f en x1
        secantes.append([p0, p1]) # Se agregan los puntos que forman la secante 

        x2 = x1 - ((x1-x0) / (p1[1]-p0[1]))*p1[1] # Siguiente iteracion

        if abs(x2 - x1) < stop or i > maxiter:
            break
            
        else:
            x0 = x1
            x1 = x2
            continue
    
    return secantes
```

```python
# EJEMPLO1---------------------------------

# Criterios de paro
detencion = 10**-3
maxiteraciones = 10

#Selecciono dos puntos iniciales
inicial0 = 1 
inicial1 = 2

ejemplo1 = secantes(f, inicial0, inicial1, detencion, maxiteraciones)

for s in ejemplo1:
    print(s)

print(len(ejemplo1))
```
